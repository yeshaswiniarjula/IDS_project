```python
import pandas as pd
import numpy as np
df = pd.read_csv("Wednesday-workingHours.pcap_ISCX.csv")
print("Original Shape:", df.shape)
# Clean column names
df.columns = df.columns.str.strip()
# 2. Data Cleaning
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
print("After Cleaning Shape:", df.shape)
# 3. Check Label Distribution
print("\nLabel Counts:")
print(df['Label'].value_counts())
# 4. Convert Label to Binary
# BENIGN = 0, Attack = 1
df['Label'] = df['Label'].apply(lambda x: 0 if x == "BENIGN" else 1)
# 5. Separate Features and Target
X = df.drop('Label', axis=1)
y = df['Label']
# 6. Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# 7. Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=42)
print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
# 8. Train Random Forest Model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
# 9. Predictions
y_pred = model.predict(X_test)
# Show first 10 predictions
print("\nFirst 10 Predictions:")
for i in y_pred[:10]:
    if i == 1:
        print("YES - Attack")
    else:
        print("NO - Normal")
# 10. Evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print("\nModel Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("\nConfusion Matrix:\n", cm)
print("\nFull Classification Report:\n")
print(classification_report(y_test, y_pred))
```

recall method 1

```python
# Get probabilities
y_prob = model.predict_proba(X_test)[:, 1]

# Lower threshold
threshold = 0.3
y_pred = (y_prob > threshold).astype(int)

print("Recall:", recall_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

maximam recall method 2

```python
model = RandomForestClassifier(
    n_estimators=300,
    class_weight={0:1, 1:7},
    random_state=42,
    n_jobs=-1
)
model.fit(X_train, y_train)
y_prob = model.predict_proba(X_test)[:, 1]
threshold = 0.3
y_pred = (y_prob > threshold).astype(int)
print("Recall:", recall_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

**using SVM**

```python
import pandas as pd
import numpy as np
# Load Dataset
df = pd.read_csv("Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv")

print("Original Shape:", df.shape)

# Remove extra spaces in column names
df.columns = df.columns.str.strip()

# Remove null values
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

print("After Cleaning Shape:", df.shape)

# Encode Label column
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Label'] = le.fit_transform(df['Label'])

# Convert BENIGN to 0 and all attacks to 1
df['Label'] = df['Label'].apply(lambda x: 0 if x == 0 else 1)

print("Label Counts:\n", df['Label'].value_counts())

# Split Features and Target
X = df.drop('Label', axis=1)
y = df['Label']

# Feature Scaling (VERY IMPORTANT for SVM)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)

# SVM MODEL

from sklearn.svm import SVC

model = SVC(kernel='rbf')   # You can try 'linear' also
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

print("First 10 Predictions:")
for i in y_pred[:10]:
    if i == 1:
        print("YES - Attack")
    else:
        print("NO - Normal")

# Accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("SVM Model Accuracy:", accuracy)

# PCA (Dimensionality Reduction)

from sklearn.decomposition import PCA

pca = PCA(n_components=20)
X_pca = pca.fit_transform(X_scaled)

print("Original Shape:", X_scaled.shape)
print("After PCA:", X_pca.shape)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)

# Recall
recall = recall_score(y_test, y_pred)
print("Recall:", recall)

# F1 Score
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Full Classification Report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Train SVM with higher weight for attacks
svm_model = SVC(kernel='rbf', class_weight={0:1, 1:3}, probability=True)
svm_model.fit(X_train, y_train)

# Get probabilities
y_prob = svm_model.predict_proba(X_test)[:,1]

# Lower threshold to increase recall
y_pred = (y_prob > 0.4).astype(int)

# Evaluation
print("SVM Results:")
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

BENIGN

Heartbleed

👉 These are mostly DoS (Denial of Service) attacks.
🔎 What’s in Friday (Afternoon – DDoS)?

BENIGN
DDoS

👉 This is a Distributed Denial of Service attack.
⚡ So Are They Similar?
✅ YES — They are similar in attack goal

Both DoS and DDoS:
Overload the server
Flood traffic
Make services unavailable
Target availability (CIA triad)
